### Data Merging

```python
df1 = pd.DataFrame({
    "cust_id": ["C1", "C2", "C3", "C4", "C5"],
    "item_bought": [10, 15, 20, 35, 4],
    "city": ["Bandung", "Jakarta", "Bogor", "Tangerang", "Bali"]
})
df1
```

```python
df2 = pd.DataFrame({
    "cust_no": ["C1", "C2", "C3", "C5", "C6", "C7", "C8"],
    "age": [24, 31, 23, 40, 35, 28, 34]
})
df2
```

```python
df1.merge(df2, how="left", left_on="cust_id", right_on="cust_no")
```

```python
df1.merge(df2, how="right", left_on="cust_id", right_on="cust_no")
```

```python
df1.merge(df2, how="inner", left_on="cust_id", right_on="cust_no")
```

```python
df1.merge(df2, how="inner", left_on="cust_id", right_on="cust_no").drop(columns=["cust_no"])
```

```python
df1.merge(df2, how="outer", left_on="cust_id", right_on="cust_no")
```

### Concat

```python
df1 = pd.DataFrame({
    "cust_id": ["C1", "C2", "C3", "C4", "C5"],
    "item_bought": [10, 15, 20, 35, 4],
    "city": ["Bandung", "Jakarta", "Bogor", "Tangerang", "Bali"]
})
df1
```

```python
df2 = pd.DataFrame({
    "cust_id": ["C6", "C7", "C8"],
    "item_bought": [12, 20, 6],
    "city": ["Bandung", "Jakarta", "Jakarta"]
})
df2
```

```python
pd.concat([df1, df2]).reset_index(drop=True)
```

### Method Apply

```python
df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [10, 20, 30, 40],
    "C": [20, 40, 60, 80]
})
df
```

#### Apply

```python
def sum(x):
    result = x.sum()
    return result
```

```python
df["sum"] = df.apply(sum, axis=1)
df
```

```python
df.loc[4,:] = df.apply(sum, axis=0)
df
```

```python
def multiply_by_2(x):
    return x*2
```

```python
df["Bx2"] = df["B"].apply(multiply_by_2)
```

```python
df
```

#### Apply without def function()

```python
# reset data
df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [10, 20, 30, 40],
    "C": [20, 40, 60, 80]
})
df
```

```python
df["sum"] = df.apply(lambda x:x.sum(), axis=1)
df
```

```python
df.loc[4,:] = df.apply(lambda x:x.sum(), axis=0)
df
```

```python
df["Bx2"] = df["B"].apply(lambda x: x*2)
df
```

### Method ApplyMap or Map

```python
# reset data
df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [10, 20, 30, 40],
    "C": [20, 40, 60, 80]
})
df
```

```python
df.applymap(np.square)
```

```python
df.map(np.square)
```
